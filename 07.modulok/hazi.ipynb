```python
# 2. valtozat kidolgozasa (tobbe-kevesbe)

def get_text():
    return 'alma alma piros alma barack'

def get_format():
    #return 'tsv'
    return 'xml'

A = ['nagy', 'kicsi', 'piros']
N = ['kutya', 'macska', 'alma']
V = ['eszik', 'iszik']

def elemzo(text):
    text = text.lower()
    text = text.split()
    res = []
    for word in text:
        if word in A:
            res.append((word, 'A'))
        elif word in V:
            res.append((word, 'V'))
        elif word in N:
            res.append((word, 'N'))
        else:
            res.append((word, ''))
    return res

def kiiro(elemzett, formatum):
    if formatum == 'tsv':
        for word, pos in elemzett:
            if pos:
                print(word, pos, sep='\t')
            else:
                print(word, '-', sep='\t')
    elif formatum == 'xml':
        xml = []
        for word, pos in elemzett:
            #if len(i[1]) > 0:
            if pos:
                xml.append('<{1}>{0}</{1}>'.format(word, pos))
            else:
                xml.append(word)
        print(' '.join(xml))
    else:
        pass

##########################

t = get_text()
f = get_format()
elemzett = elemzo(t)
kiiro(elemzett, f)
```
